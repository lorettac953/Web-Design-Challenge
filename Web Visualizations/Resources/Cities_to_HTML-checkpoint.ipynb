```python
# Dependencies
import os
import pandas as pd
```

```python
file_path = os.path.join('Resources/cities.csv')
cities = pd.read_csv(file_path)
```

```python
cities_df = pd.DataFrame(cities)
```

```python
cities_df
```

```python
cities = cities_df.reset_index(drop=True)
cities.head()
```

```python
cities_data = cities.rename(columns = {"City_ID":"City ID",
                                      "City":"City Name",
                                      "Cloudiness":"Cloudiness",
                                      "County":"County",
                                      "Date":"Date",
                                      "Humidity":"Humidity %",
                                      "Lat":"Latitude",
                                      "Lng":"Longitude",
                                      "Max Temp":"Max Temperature(F)",
                                      "Wind Speed":"Wind Speed(MPH)"})
cities_data
```

```python
cities_data["Date"] = pd.to_datetime(cities_data['Date'])
cities_data
```

```python
cities_data.to_html('Resources/data.html')
```
